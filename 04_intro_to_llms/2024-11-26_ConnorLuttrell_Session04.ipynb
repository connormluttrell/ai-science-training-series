```python
from transformers import AutoTokenizer

# A utility function to tokenize a sequence and print out some information about it.

def tokenization_summary(tokenizer, sequence):

    # get the vocabulary
    vocab = tokenizer.vocab
    # Number of entries to print
    n = 10

    # Print subset of the vocabulary
    print("Subset of tokenizer.vocab:")
    for i, (token, index) in enumerate(tokenizer.vocab.items()):
        print(f"{token}: {index}")
        if i >= n - 1:
            break

    print("Vocab size of the tokenizer = ", len(vocab))
    print("------------------------------------------")

    # .tokenize chunks the existing sequence into different tokens based on the rules and vocab of the tokenizer.
    tokens = tokenizer.tokenize(sequence)
    print("Tokens : ", tokens)
    print("------------------------------------------")

    # .convert_tokens_to_ids or .encode or .tokenize converts the tokens to their corresponding numerical representation.
    #  .convert_tokens_to_ids has a 1-1 mapping between tokens and numerical representation
    # ids = tokenizer.convert_tokens_to_ids(tokens)
    # print("encoded Ids: ", ids)

    # .encode also adds additional information like Start of sequence tokens and End of sequene
    print("tokenized sequence : ", tokenizer.encode(sequence))

    # .tokenizer has additional information about attention_mask.
    # encode = tokenizer(sequence)
    # print("Encode sequence : ", encode)
    # print("------------------------------------------")

    # .decode decodes the ids to raw text
    ids = tokenizer.convert_tokens_to_ids(tokens)
    decode = tokenizer.decode(ids)
    print("Decode sequence : ", decode)


tokenizer_1  =  AutoTokenizer.from_pretrained("gpt2") # GPT-2 uses "Byte-Pair Encoding (BPE)"

sequence = "Counselor, please adjust your Zoom filter to appear as a human, rather than as a cat"

tokenization_summary(tokenizer_1, sequence)
```

```python
# STEP 0 : Installations and imports
from transformers import AutoTokenizer, AutoModelForSequenceClassification, AutoConfig
import torch
import torch.nn.functional as F
```

```python
# STEP 1 : Set up the prompt
input_text = "The panoramic view of the ocean was breathtaking."
```

```python
# STEP 2 : Load the pretrained model.
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
config = AutoConfig.from_pretrained(model_name)
print(config)
```

```python
#STEP 3 : Load the tokenizer and tokenize the input text
tokenizer  =  AutoTokenizer.from_pretrained(model_name)
input_ids = tokenizer(input_text, return_tensors="pt")["input_ids"]
print(input_ids)
```

```python
# STEP 5 : Perform inference
outputs = model(input_ids)
result = outputs.logits
print(result)

# STEP 6 :  Interpret the output.
probabilities = F.softmax(result, dim=-1)
print(probabilities)
predicted_class = torch.argmax(probabilities, dim=-1).item()
labels = ["NEGATIVE", "POSITIVE"]
out_string = "[{'label': '" + str(labels[predicted_class]) + "', 'score': " + str(probabilities[0][predicted_class].tolist()) + "}]"
print(out_string)
```

# Homework, Part 1
* The higher temperatures of GPT-2 and OPT-1.3B leads to varying and less structured responses, but prioritizes creativity.
* The lower/balanced temperatures of Phi-3-mini-4k and Blooms prioritize coherence. This produces a more logical output.
* The shorter limits of max new tokens potentially cuts off thoughts, constraining the responses.
* The longer limtis of max new tokens enables more detail with a downfall of topic drift.
* The larget models have a better understanding of context, enabling them to produce more coherent text.
* The smaller models are known to struggle in coherence maintenance with longer outputs.

# Homework, Part 2
* OPT-1.3B is a larger model that has more attention heads and layers.
* This enables the model to capture better dependencies and distribute attention better across the sequence.
* GPT-2 is a smaller model that relies on basic localized attention patterns.
* It often focuses on local tokens and the model struggles with nuanced relations.
* Larger models provide specialization in sentence structure and smaller models show repetitive patterns.
* Larger models are considered to be more contextually aware but generally slower.

```python
# Install BertViz for visualizing attention mechanisms
# !pip install bertviz

from transformers import AutoTokenizer, AutoModelForCausalLM, utils
from bertviz import model_view
utils.logging.set_verbosity_error()  # Suppress standard warnings

# Define input text
input_text = "Forget it Jake, it's Chinatown."

# Model 1: GPT2
model_gpt2 = AutoModelForCausalLM.from_pretrained("gpt2", output_attentions=True)
tokenizer_gpt2 = AutoTokenizer.from_pretrained("gpt2")
inputs_gpt2 = tokenizer_gpt2.encode(input_text, return_tensors="pt")
outputs_gpt2 = model_gpt2(inputs_gpt2)
attention_gpt2 = outputs_gpt2[-1]
tokens_gpt2 = tokenizer_gpt2.convert_ids_to_tokens(inputs_gpt2[0])

# Visualize GPT2 attention
print("### GPT2 Attention Visualization ###")
model_view(attention_gpt2, tokens_gpt2)

# Model 2: facebook/opt-1.3b
model_opt = AutoModelForCausalLM.from_pretrained("facebook/opt-1.3b", output_attentions=True)
tokenizer_opt = AutoTokenizer.from_pretrained("facebook/opt-1.3b")
inputs_opt = tokenizer_opt.encode(input_text, return_tensors="pt")
outputs_opt = model_opt(inputs_opt)
attention_opt = outputs_opt[-1]
tokens_opt = tokenizer_opt.convert_ids_to_tokens(inputs_opt[0])

# Visualize OPT-1.3B attention
print("### OPT-1.3B Attention Visualization ###")
model_view(attention_opt, tokens_opt)
```
